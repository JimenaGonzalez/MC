```python
#import numpy as np
#import matplotlib as mpl
#import matplotlib.pyplot as plt
#import os
#%matplotlib inline
#from IPython.display import HTML
%pylab inline
from scipy import interpolate
```

```python
def sin(x):
    return np.sin(x)
x= (0, np.pi/2 , np.pi, (np.pi)*3/2 , 2*np.pi)
xmas= linspace(0, 2*np.pi, 8)
#Puntos de muestra
y= sin(x)
ymas=sin(xmas)
plot(x,y)
```

```python
#Lagrange:
def lagrangep(absc,j):
    pol=np.poly1d([0,1])
    k=len(absc)
    for m in range(k):
        if m!=j:
            pol*=1./(absc[j]-absc[m])*np.poly1d([1.,-absc[m]])
    return pol 
def interlagr(absc,orde):
    poly=np.poly1d([0])
    for i in range(len(absc)):
        poly+=orde[i]*lagrangep(absc,i)
    return poly
po1=interlagr(x,y)
```

```python
#Grafica de los puntos vs la interpolacion de Lagrange
xcont=np.linspace(0,2*np.pi,100)
ycontl1= po1(xcont)
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(xcont,ycontl1)
ax.plot(x,y,"o", label= "Puntos de muestra")
#plt.legend()
#plt.title("Interpolación de Lagrange\n",fontsize=20)
#plt.show(fig)
```

```python
#Grafica de los puntos vs la interpolacion de Lagrange para mas puntos
po2=interlagr(xmas,ymas)
ycontl2= po2(xcont)
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(xcont,ycontl2)
ax.plot(x,y,"o", label= "Puntos de muestra")
#plt.legend()
#plt.title("Interpolación de Lagrange\n",fontsize=20)
#plt.show(fig)
```

```python
#Linear spline
interpol_lin = interpolate.interp1d(x,y,kind='linear')
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(xcont,interpol_lin(xcont),label="linear")
ax.plot(x,y,"o", label= "Puntos de muestra")
```

```python
#Interpolacion lineal con mas puntos
interpol_lin2 = interpolate.interp1d(xmas,ymas,kind='linear')
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(xcont,interpol_lin2(xcont),label="linear")
ax.plot(x,y,"o", label= "Puntos de muestra")
```

```python
#Cubic spline
interpol_cubic = interpolate.interp1d(x,y,kind='cubic')
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(xcont,interpol_cubic(xcont),label="cubic")
ax.plot(x,y,"o", label= "Puntos de muestra")
```

```python
#Cubic spline con mas puntos
interpol_cubic2 = interpolate.interp1d(xmas,ymas,kind='cubic')
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(xcont,interpol_cubic2(xcont),label="cubic")
ax.plot(x,y,"o", label= "Puntos de muestra")
```
