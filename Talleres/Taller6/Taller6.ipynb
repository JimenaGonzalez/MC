```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
%matplotlib inline
from scipy.fftpack import ifft, fft, fftfreq
from scipy import signal
from numpy import zeros
from scipy import interpolate
```

```python
data=np.genfromtxt("a.csv", delimiter=",")
phi= np.pi/2 - np.radians(data[:,1]) #latitud
theta=np.radians(data[:,0]) #longitud
#tiempo en horas
t= data[:,3] + data[:,4]/60 + data[:,5]/3600
#radio tierra km
R=6378
```

```python
x= R*np.sin(phi)*np.cos(theta)
y= R*np.sin(phi)*np.sin(theta)
z= R*np.cos(phi)
```

```python
#Vectores de desplazamiento:
xn=x - x[0]
yn=y - y[0]
zn=z - z[0]
tn=t - t[0]
```

```python
#Punto de referencia: latitud 4.5981, longitud -74.0758.
#REFERENCIA
rpphi= np.pi/2 - np.radians(4.5981)
rptheta=np.radians(-74.0758)
```

```python
unitariooriente=[-np.sin(rptheta),np.cos(rptheta),0]
unitarionorte=[-np.cos(rptheta)*np.cos(rpphi),-np.sin(rptheta)*np.cos(rpphi) ,np.sin(rpphi) ]
localx=np.zeros(len(xn))
localy=np.zeros(len(xn))
for i in range (len(xn)):
    r= np.array([xn[i],yn[i],zn[i]])
    localx[i]= np.sum (r * unitariooriente)
    localy[i]= np.sum (r * unitarionorte)
```

```python
plt.figure(figsize=(4,10))
plt.plot(localx,localy)
plt.show()
#smopy.map
```

```python
localx=localx-np.roll(localx, 1)
localx[0]=0.
localy=localy-np.roll(localy, 1)
localy[0]=0.
dr=np.sqrt(localx**2+localy**2)
dracu=np.cumsum(dr)

# Usando el metodo del trapecio, no lo grafique porque no me dio
rm= np.sqrt(localx**2 + localy**2)
def trapz_slow(x, y):
    area = 0.
    for i in range(1, len(x)):
        area += (x[i] - x[i-1]) * (y[i] + y[i-1])
    return area / 2
distan=[]
for i in range(len(tn)):
    distan.append(trapz_slow(tn[0:i],rm[0:i]))

plt.figure(figsize=(10,6))
with plt.xkcd():
    plt.plot(tn,dracu)
    plt.title("A walk in the park")
    plt.xlabel("t/h")
    plt.ylabel("d/km")
```
