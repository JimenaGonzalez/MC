```python
#Importar librerías 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
from numpy import zeros
from scipy.fftpack import ifft, fft, fftfreq, fft2, ifft2, fftn, ifftn
%matplotlib inline
from scipy import interpolate
from scipy.optimize import fminbound
```

## Punto a

```python
#Se creó un arreglo de 1024 x 1024 de ceros y a las casillas en el centro del arreglo (en un cuadrado de 32) se les asignó otro valor
arreglo = zeros((1024,1024), int)
for i in range(495,527):
    for j in range(495,527):
        arreglo[i][j]=200
```

## Punto b

```python
#Se aplicó la transformada de Fourier al arreglo 
arreglo_dft=fft2(arreglo)
#Se hizo un cambio de coordenadas
arreglo_nuevo= np.fft.fftshift(arreglo_dft)
#Se tomo del centro de arreglo_nuevo un cuadrado de lado 200 (411:611)
cuadro = zeros((200,200), float)
for i in range(200):
        for j in range(200):
            cuadro[i][j]= np.abs(arreglo_nuevo[i+410][j+410])           
#Se resalta usando log
mag_cuadro=np.log((np.abs(cuadro))**2)
```

```python
#Se crea la figura y se agregan las dos gráficas hechas antes
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.imshow(1.5*arreglo,cmap=plt.cm.gray)
plt.subplot(1,2,0)
plt.imshow(1.5*mag_cuadro)
plt.title(u'Patrón de interferencia de una apertura cuadrada \n (log resaltado)')
plt.show()
```

## Punto d

```python
#Se crea un arreglo 1024 x 1024 y se inicializa en cero
circulo = zeros((1024,1024), int)
#A los puntos que se encuentren dentro del circulo ( x**2+y**2 <= r**2 ) se les asigna otro valor
for i in range(0,1024):
    for j in range(0,1024):
        if( np.sqrt( (i-512)**2 + (j-512)**2 ) <= 10.24):          
            circulo[i][j]=1    
```

```python
#Se aplica la transformada de Fourier
circulo_dft=fft2(circulo)
#Se hace un cambio de coordenadas
circulo_nuevo= np.fft.fftshift(circulo_dft)
#Se resalta con log
mag_circulo_nuevo=np.log((np.abs(circulo_nuevo)+1)**2)
```

```python
# Para realizar la gráfica superior (intesidad vs posición)
#Se crean dos lista para guardas los valores de las abcisas y las coordenadas
xc=[]
yc=[]
#Se asume que la amplitud máxima (intensidad máxima) se encuentra en el centro del arreglo de circulo_nuevo
maxa=circulo_nuevo[512][512]
#Para hallar los valores de las coordenadas se elige toda la fila central del arreglo (circulo_nuevo[512,:]) se encuentra su norma
#y se eleva al cuadrado, se divide entre el la norma del valor máximo de intensidad (maxa) al cuadrado (para normalizar) 
yc= (abs(circulo_nuevo[512,:])**2)/abs(maxa)**2
#Se toman los valores para las abcisas de esta manera para poder comparar estos resultados con la teoría facilmente.
xc= 2*np.pi*0.01*np.linspace(-512,512,1024)
```

```python
#Se crea la figura y se define su tamaño
plt.figure(figsize=(15,10))
plt.subplot2grid((3,2),(0,0),colspan=2,rowspan=1)
#Se grafica intensidad vs posición
plt.plot(xc,yc,"o")
#Se limita el dominio de x
plt.xlim(0,10)
#Se asignan los títulos de los ejes
plt.ylabel(r"$I/I_0$")
plt.xlabel(r"$2 \pi R j/L$")
plt.subplot2grid((3,2),(1,0),rowspan=2)
plt.imshow(1.5*circulo,cmap=plt.cm.gray)
plt.subplot2grid((3,2),(1,1),rowspan=2)
plt.imshow(1.5*mag_circulo_nuevo)
plt.title(u'Patrón de interferencia de una apertura circular (Patrón de Airy) \n log resaltado')
plt.tight_layout()
plt.show()
```

## Punto e

```python
# Puntos escogidos para hacer la interpolación, se toman la mitad de los puntos dados de manera uniforme

xp=[]
yp=[]
for i in range(512):
    xp.append(xc[i*2])
    yp.append(yc[i*2])
#Se hace una interpolación cúbica con los puntos escogidos
interpol_cubic = interpolate.interp1d(xp,yp,kind='cubic')
#Se crea una lista con los intervalos entre los cuales están máximos y mínimos esperados.
inter=[-1,3,4,6,8,9,11,12,13]
#Se crea una lista minimos para guardar los mínimos encontrados, y mivalor para guardar el valor del mínimo correspondiente
minimos=[]
mivalor=[]
#Se utiliza el comando fminbound para encontrar el punto en x donde se encuentra un mínimo, a este comando se le da como parámetro
#la función , y el rango donde es posible encontrar el mínimo (se uso la lista inter que tiene estos rangos)
for i in range (3):
    minimos.append(fminbound(interpol_cubic,inter[i*2+1],inter[i*2+3]))
    #Se usa la interpolacion para encontrar el valor del mínimo
    mivalor.append(float(interpol_cubic(minimos[i])))
#Debido a que no hay una función equivalente a fminbound para hallar máximos lo que se hace es reflejar la función sobre el eje x
#Esta nueva función se determina así: se toman los puntos anteriores escogidos para y (yp) y se reflejan sobre el eje x (se multi-
#plican por -1 y se guardan en una nueva lista ypm
ypm=[]
for i in range(len(yp)):
    ypm.append(-(yp[i]))
#Se encuentra una interpolación cúbica para estos puntos
interpol_cubic_max = interpolate.interp1d(xp,ypm,kind='cubic')
#Se crean dos lista, maximos para guardar el punto donde se ubica el máximo y mavalor para guardar el valor de éste
maximos=[]
mavalor=[]
for i in range (4):              
    maximos.append(fminbound(interpol_cubic_max,inter[i*2],inter[i*2+2]))
    mavalor.append(float(interpol_cubic(maximos[i])))
```

```python
#En una sola lista se guardan máximo y mínimos y en otra, se guardan los valores correspondientes.
puntosx=[]
puntosy=[]
for i in range (len(minimos)):
    puntosx.append(maximos[i])
    puntosy.append(mavalor[i])
    puntosx.append(minimos[i])
    puntosy.append(mivalor[i])
puntosx.append(maximos[3])
puntosy.append(mavalor[3])
```

```python
#Puntos teóricos para hacer la comparación
xt= [0,3.833,5.136,7.016,8.417,10.174,11.620]
yt= [1,0,0.0175,0,0.0042,0,0.0016]

for i in range(len(xt)):
    if (i%2==0):
        print "Hay un máximo en " + str(puntosx[i]) + " y su valor es " + str(puntosy[i])
        print "Según la teoría el máximo se encuentra en " + str(xt[i]) + " y su valor es " + str(yt[i]) + "\n"
        
    if (i%2!=0):
        print "Hay un mínimo en " + str(puntosx[i]) + " y su valor es " + str(puntosy[i])
        print "Según la teoría el mínimo se encuentra en " + str(xt[i]) + " y su valor es " + str(yt[i]) + "\n"
 
```

```python
plt.figure(figsize=(10,10))
plt.plot(puntosx,puntosy,"o",ms=10, label= "Puntos encontrados")
plt.plot(xt,yt,"o",ms=10,label=u"Puntos teóricos")
plt.legend()
plt.ylabel("I/I_0")
plt.xlabel("2 pi R j/L")
plt.xlim(-0.1,12)
plt.ylim(-0.2,1.2)
plt.show()
#En la gráfica se muestran los puntos esperados y los puntos hallados, y se observa que los valores son muy parecidos. Varios puntos
#se superponen.
```
