```python
#Se importan las librerias necesarias
%pylab inline
import turtle
from IPython.display import Image
import matplotlib.image
```

```python
#Longitud de cada lado del triangulo inicial
long=500
#se establece el tamaño de la pantalla
turtle.setup(800, 800)     
#se le da nombre a la pantalla 
wn = turtle.Screen() 
# se elige el color de la pantalla
wn.bgcolor("white")
# se inicializan tres tortugas
t1=turtle.Turtle()
t2=turtle.Turtle()
t3=turtle.Turtle()
# se selecciona la mayor rapidez para cada tortuga
t1.speed(0)
t2.speed(0)
t3.speed(0)
#se eligen las posiciones iniciales de las tortugas, se eligen de manera que cuando se dibuje el triangulo
# este quede centrado, la tortuga 1 a la izquierda y abajo, la dos derecha y abajo y la 3 arriba y en el centro
t1.goto(-250,-200)
t2.goto(250,-200)
t2.left(120)
t3.goto(250,-200)
t3.left(120)
t3.forward(500)
t3.left(120)
#se eliminan los trazos que hicieron para llegar a sus posiciones iniciales
t1.clear()
t2.clear()
t3.clear()
#se esconden las tortugas
t1.hideturtle()
t2.hideturtle()
t3.hideturtle()
#se elige el tamaño de la figura
size = 500
#Definición de la función recursiva, le entra como paramétro las 3 tortugas, el número del paso y la longitud de este
def copo1(t1,t2,t3,n, long):
    #vuelvo a cambiar el tamaño de la figura
    size = 50
    # paso base, cada tortuga avanza una distancia long
    if (n==0):
        t1.forward(long)
        t2.forward(long)
        t3.forward(long)
    else:
        #Para los otros pasos se vuelve a llamar a la función disminuyendo en 1 el paso correspondiente, hasta que 
        #llegue al caso base, la longitud disminuye a long/3.
            copo1(t1,t2,t3,n-1,long/3)
            #se cambia la orientacion de las tortugas para que esten listas para el siguiente movimiento
            t1.left(-60)
            t2.left(-60)
            t3.left(-60)
            # se reajusta el tamaño
            size = size*3
            copo1(t1,t2,t3,n-1,long/3)
            t1.left(120)
            t2.left(120)
            t3.left(120)
            size = size*3
            copo1(t1,t2,t3,n-1,long/3)
            t1.left(-60)
            t2.left(-60)
            t3.left(-60)
            size = size*3
            copo1(t1,t2,t3,n-1,long/3)

#Esta función gráfica el paso correspondiente (en este caso el número 0)
copo1(t1,t2,t3, 1, 500)
#se guarda la imagen generada por turtle en un archivo .eps
turtle.getcanvas().postscript(file = "1.eps")
wn.exitonclick()
```

Para guardar las 6 imagenes se implemento la funcion copo1 6 veces, cambiando solo el numero del paso (de 0 a 5), 
cada imagen se guardo en un archivo .eps y desde la terminal se cambio el formato de cada imagen a png usando el
siguiente comando: `convert 0.eps 0.png`, y para hacer la animación se usó el comando: ` convert -delay 30 -loop 0 *.eps animacion.gif
`.

```python
# se define la figura
figura = figure(figsize=(20,20)) 
#se hace un ciclo (son 6 imagenes)
for i in range(6):
    #se  agregan subgraficas a la figura
    figura.add_subplot(2,3,i+1) 
    #convercion de int a string
    num= str(i)
    #se elige la imagen a mostrar
    imagen = imread(num + '.png')
    #se coloca un título a cada imagen
    title ("Paso " + num)
    #se muestra cada imagen
    imshow(imagen)
    #se ocultan los ejes
    axis ('off')
#se guarda la imagen
savefig("panel.png")
```
